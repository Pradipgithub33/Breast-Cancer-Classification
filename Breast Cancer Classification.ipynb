```python
X = df.drop('diagnosis', axis = 1)
Y = df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.3, random_state= 33)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, roc_auc_score,confusion_matrix , classification_report ,roc_curve,auc
#Fitting the model
xgb=XGBClassifier()
xgb.fit(x_train, y_train)

#Prediction on test data
y_pred1 = xgb.predict(x_test)

#Prediction on train data
y_pred_train1 = xgb.predict(x_train)

#Model Evaluation
#Print accuracy and Roc auc values of XGB model
print("Train Accuracy of XGB model is : ", accuracy_score(y_train,y_pred_train1))
print("Test Accuracy of XGB model is : ", accuracy_score(y_test,y_pred1))
print("ROC_AUC Score of XGB model is :",roc_auc_score(y_pred1,y_test))
print(classification_report(y_pred1,y_test))

#Confusion matrix 
con_mat = confusion_matrix(y_test, y_pred1)
print("Confusion matrix of XGB model is :",con_mat )
sns.heatmap(con_mat,annot=True)
```

```python
from sklearn.ensemble import RandomForestClassifier

#Fitting the model
RFC =RandomForestClassifier()
RFC.fit(x_train, y_train)

#Prediction on test data
y_pred2 = RFC.predict(x_test)

#Prediction on train data
y_pred_train2 = RFC.predict(x_train)

#Model Evaluation
#Print accuracy and Roc auc values of RFC model
print("Train Accuracy of RFC model is : ", accuracy_score(y_train,y_pred_train2))
print("Test Accuracy of RFC model is : ", accuracy_score(y_test,y_pred2))
print("ROC_AUC Score of RFC model is :",roc_auc_score(y_pred2,y_test))
print(classification_report(y_pred2,y_test))

#Confusion matrix 
con_mat = confusion_matrix(y_test, y_pred2)
print("Confusion matrix of XGB model is :",con_mat )
sns.heatmap(con_mat,annot=True)
```
